```python
def avg(scores_list):
    total = sum(scores_list)
    average = total / len(scores_list)
    return average


name = (input("Enter your name: "))
nuofscore = int(input("Enter the number of quiz scores: "))
scores = []
for i in range(nuofscore):
    enterscore = int(input(f"Enter quiz score {i+1}: "))

# Calculate and print the average score
average_score = avg(scores)
print(f"The average score for {name} is: {average_score:.2f}")

# Determine if the average score is 'Passed' or 'Failed'
passing_threshold = 70 # You can change this value
if average_score >= passing_threshold:
    print(f"{name} has Passed with an average of {average_score:.2f}")
else:
    print(f"{name} has Failed with an average of {average_score:.2f}")
```
